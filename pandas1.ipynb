```python
dataframe2=pd.read_csv('c/deliveries.csv')

```

```python
mask_six=dataframe2['batsman_runs']==6
six=dataframe2[mask_six]
heatmapx=six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')

heatmapx.fillna(0,inplace=True)
```

```python
import seaborn as sns

sns.heatmap(heatmapx,cmap='viridis')
```
